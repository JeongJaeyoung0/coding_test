### 1부터 n 사이에 있는 소수의 개수를 반환

```python
def solution(n):
    p=[*range(n+1)]
    for i in p:
        if 1<i:p[i*i::i]=-(i+~n//i)*[0]
        if 1<i:print(i)
    return
```

```python
solution(10)
```

```python
solution(5)
```
