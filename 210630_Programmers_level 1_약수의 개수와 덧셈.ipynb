#programmers_level 1_약수의 개수와 덧셈
### left부터 right까지 모든 정수들 중 약수의 개수가 짝수는 더하고, 홀수는 뺀 수를 출력

```python
def solution(left, right):
    return sum([i if len([j for j in range(1,i+1) if i%j==0])%2 == 0 else -i for i in range(left, right+1) ])
```

```python
solution(13, 17)
```

```python
solution(24, 27)
```
