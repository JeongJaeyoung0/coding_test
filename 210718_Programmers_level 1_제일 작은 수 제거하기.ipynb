#programmers_level 1_제일 작은 수 제거하기
### 정수 배열 arr 에서 가장 작은 수를 제거한 배열을 반환. 단, 리턴하려는 배열이 빈 배열인 경우엔 배열에 -1을 채워 반환.

```python
def solution(arr):
    arr.remove(min(arr))    # 리스트에서 최소값 제거
    if len(arr) == 0:       # 제거한 리스트가 빈 리스트일 경우
        return [-1]         # [-1] 반환
    return arr              # 최소값 제거한 리스트 반환
```

```python
solution([4,3,2,1])
```

```python
solution([10])
```
