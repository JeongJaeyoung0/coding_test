### 단어 s의 가운데 글자를 반환. 단, 짝수면 가운데 두글자를 반환.

```python
import math
def solution(s):
    mid = math.ceil(len(s)/2-1)     # 중앙값
    if len(s)%2: answer = s[mid]    # 길이가 짝수일 경우
    else: answer = s[mid:mid+2]     # 길이가 홀수일 경우
    return answer
```

```python
solution("abcde")
```

```python
solution("qwer")
```
