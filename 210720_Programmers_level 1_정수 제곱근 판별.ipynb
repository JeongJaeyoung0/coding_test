#programmers_level 1_정수 제곱근 판별
### 양의 정수 n에 대해, n이 어떤 양의 정수 x의 제곱인지 아닌지 판단하여 제곱이라면 x+1의 제곱을 반환하고, 아니라면 -1을 반환.

```python
def solution(n):
    root = n**0.5                                       # 루트n
    return (root+1)**2 if str(root)[-1] == '0' else -1  # 소수점이 0이면 제곱근으로 판별
```

```python
solution(121)
```

```python
solution(3)
```
