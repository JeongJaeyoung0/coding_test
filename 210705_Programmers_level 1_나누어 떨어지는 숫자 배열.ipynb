### array % divisor == 0 인 값을 오름차순으로 정렬한 배열을 반환. 단, 요소가 없을 경우 -1을 담아 반환

```python
def solution(arr, divisor):
    return sorted([i for i in arr if n%divisor==0]) or [-1]   # 앞의 값이 거짓(0) 이면 or 뒤의 값이 반환됨
```
