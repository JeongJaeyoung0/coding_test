### 자연수 n을 뒤집어 각 자리 숫자를 원소로 가지는 배열 형태로 반환

```python
# for문
def solution(n):
    return [int(i) for i in str(n)[::-1]]   # 문자열 > for문 > 정수열 > 리스트 > 반환
```

```python
# reversed
def solution(n):
    return list(map(int, reversed(str(n)))) # 문자열 > 리버스 > map 정수열 > 리스트 > 반환
```

```python
solution(12345)
```
